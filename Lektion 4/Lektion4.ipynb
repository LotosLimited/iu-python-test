```python
# Beispieldaten — Noten zweier Gruppen
gruppe_1 = [1.7, 2.3, 1.0, 2.7, 1.3]
gruppe_2 = [3.0, 3.7, 2.7, 3.3, 4.0]

# t-Test — sind die Noten signifikant unterschiedlich?
t_wert, p_wert = stats.ttest_ind(gruppe_1, gruppe_2)

print(f"t-Wert: {t_wert}")
print(f"p-Wert: {p_wert}")

# Normalverteilung prüfen
print(stats.normaltest(gruppe_1))
```

```python
# Interpolation — fehlende Werte schätzen
x_bekannt = np.array([0, 1, 2, 3, 4, 5])
y_bekannt = np.array([0, 4, 2, 3, 7, 9])

import pandas as pd

pd.DataFrame(x_bekannt,y_bekannt)

# lineare Interpolation
linear = interp1d(x_bekannt, y_bekannt, kind="linear")
print(f"geschätzter Wert bei x=1.5 : {linear(2)}")

#kubische Interpolation
kubisch = interp1d(x_bekannt, y_bekannt, kind="cubic")
print(f"geschätzter Wert bei x=1.5 : {kubisch(2)}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# Wertebereiche generieren
n = np.arange(0, 10)
x = np.linspace(0, 9, 100)

# tatsächliche Werte
y_actual = np.cos(x)

# simulierte Messwerte mit zufälligem Rauschen
y_experiment = np.cos(n) + 0.1 * np.random.randn(len(n))


# Interpolationen berechnen
linear = interp1d(n, y_experiment, kind="linear")
kubisch = interp1d(n, y_experiment, kind="cubic")
y_linear = linear(x)
y_kubisch = kubisch(x)

# Vergleichsplot
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(n, y_experiment, "bs", label="Messdaten")
ax.plot(x, y_actual, "k", lw=2, label="Echte Funktion")
ax.plot(x, y_linear, "r", label="Lineare Interpolation")
ax.plot(x, y_kubisch, "g", label="Kubische Interpolation")
ax.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolationsvergleich")
plt.show()
```

```python
from numba import jit
import numpy as np
import time

# normale Python-Funktion
def berechne_normal(array):
    summe = 0
    for i in range(len(array)):
        summe += array[i]
    return summe

# gleiche Funktion mit Numba beschleunigt
@jit
def berechne_numba(array):
    summe = 0
    for i in range(len(array)):
        summe += array[i]
    return summe

# großen Array erstellen
grosser_array = np.random.randn(1000000)

# Zeit messen — normal
start = time.time()
berechne_normal(grosser_array)
print(f"Normal: {time.time() - start:.4f} Sekunden")

# Zeit messen — Numba
start = time.time()
berechne_numba(grosser_array)
print(f"Numba: {time.time() - start:.4f} Sekunden")
```

```python
# zweiter Aufruf — jetzt ist Numba bereits kompiliert!
start = time.time()
berechne_normal(grosser_array)
print(f"Normal: {time.time() - start:.4f} Sekunden")

start = time.time()
berechne_numba(grosser_array)
print(f"Numba: {time.time() - start:.4f} Sekunden")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# normalverteilte Daten generieren
# mu = Mittelwert, sigma = Standardabweichung
mu = 100
sigma = 15
x = mu + sigma * np.random.randn(10000)  # 10.000 zufällige Werte

fig, ax = plt.subplots(figsize=(6, 4))

# Histogramm erstellen
ax.hist(x,
        bins=50,          # Anzahl der Balken
        density=True,     # Wahrscheinlichkeitsdichte statt Anzahl
        facecolor='r',    # Balkenfarbe rot
        alpha=0.75,       # 75% Deckkraft — leicht transparent
        label="Histogramm Beispiel")

# Text-Annotation mit mathematischer Notation
ax.text(45, 0.028, r'$\mu=100,\ \sigma=15$')

# Achsenbereiche festlegen [x_min, x_max, y_min, y_max]
ax.axis([40, 160, 0, 0.03])

plt.legend()
plt.grid(True, color="k")
plt.xlabel("x")
plt.ylabel("P(x)")
plt.title("Histogramm")
plt.show()
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.models import HoverTool

teams = ["A", "B", "C", "D", "E", "F"]
score = [4, 2, 3, 1, 3, 5]

output_file("balken_hover.html")

plot = figure(x_range=teams,
              height=500,
              title="Balkendiagramm mit Hover",
              x_axis_label="Teams",
              y_axis_label="Punkte")

plot.vbar(x=teams,
          top=score,
          width=0.6,
          legend_label="Punkte")

# HoverTool hinzufügen
hover = HoverTool(tooltips=[
    ("Team", "@x"),        # zeigt den Team-Namen
    ("Punkte", "@top")     # zeigt die Punktzahl
])
plot.add_tools(hover)

show(plot)
```

```python
import sqlite3

# Verbindung zur Datenbank herstellen
# (erstellt automatisch eine neue .db Datei!)
connection = sqlite3.connect("meine_datenbank.db")

# Cursor erstellen — damit führen wir SQL-Befehle aus
cursor = connection.cursor()

# Tabelle erstellen
cursor.execute("""
    CREATE TABLE IF NOT EXISTS studenten (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        vorname TEXT NOT NULL,
        nachname TEXT NOT NULL,
        alter_student INTEGER NOT NULL,
        note REAL,
        aktiv BOOLEAN NOT NULL
    )
""")

# Änderungen speichern
connection.commit()
print("Tabelle erfolgreich erstellt!")
```

```python
#einzelnen Datensatz einügen

cursor.execute("""
               INSERT INTO studenten (vorname, nachname, alter_student, note, aktiv)
               VALUES (?,?,?,?,?)
""", ("Michelle", "Müller", 22 , 1.7, True))

# mehrere Datensätze auf einmal einfügen
studenten_liste = [
    ("Jonas", "Schmidt", 25, 2.3, True),
    ("Anna", "Weber", 23, 3.0, True),
    ("Tim", "Fischer", 28, None, False),
    ("Sara", "Meyer", 21, 1.3, True)
]

cursor.executemany("""
    INSERT INTO studenten (vorname, nachname, alter_student, note, aktiv)
    VALUES (?, ?, ?, ?, ?)
""", studenten_liste)


connection.commit()
print("Daten erfolgreich eingefügt!")
```

```python
# Tims Note aktualisieren
cursor.execute("""
    UPDATE studenten 
    SET note = 2.7
    WHERE vorname = 'Tim' AND nachname = 'Fischer'
""")

connection.commit()
print("Note aktualisiert!")

# prüfen ob es geklappt hat
cursor.execute("SELECT vorname, nachname, note FROM studenten WHERE vorname = 'Tim'")
print(cursor.fetchone())

# mehrere Spalten gleichzeitig aktualisieren
cursor.execute("""
    UPDATE studenten
    SET note = 4.0, aktiv = 0
    WHERE vorname = 'Anna'
""")

connection.commit()
print("Anna aktualisiert!")

# alle Änderungen anzeigen
cursor.execute("SELECT vorname, nachname, note, aktiv FROM studenten")
for student in cursor.fetchall():
    print(student)
```

```python
# einen Studenten löschen
cursor.execute("""
    DELETE FROM studenten
    WHERE vorname = 'Tim' AND nachname = 'Fischer'
""")

connection.commit()
print("Tim wurde gelöscht!")

# prüfen was noch in der Tabelle ist
cursor.execute("SELECT * FROM studenten")
print("Verbleibende Studenten:")
for student in cursor.fetchall():
    print(student)
```

```python
# nach einer Bedingung löschen
cursor.execute("""
    DELETE FROM studenten
    WHERE note > 3.0
""")

connection.commit()
print("Studenten mit Note schlechter als 3.0 gelöscht!")

# finaler Stand
cursor.execute("SELECT * FROM studenten")
print("Finaler Stand:")
for student in cursor.fetchall():
    print(student)
```

```python
# ORDER BY — Ergebnisse sortieren
cursor.execute("""
    SELECT vorname, nachname, note 
    FROM studenten 
    ORDER BY note ASC
""")
print("Nach Note sortiert (aufsteigend):")
for student in cursor.fetchall():
    print(student)

print("---")
```

```python
# AVG — Durchschnitt berechnen
cursor.execute("""
    SELECT AVG(note) 
    FROM studenten 
    WHERE note IS NOT NULL
""")
print(f"Durchschnittsnote: {cursor.fetchone()[0]}")
```

```python
# COUNT — Anzahl zählen
cursor.execute("SELECT COUNT(*) FROM studenten")
print(f"Anzahl Studenten: {cursor.fetchone()[0]}")
```

```python
cursor.execute("""
    SELECT vorname, nachname 
    FROM studenten 
    WHERE vorname LIKE 'M%'
""")
print("Studenten deren Name mit M beginnt:")
for student in cursor.fetchall():
    print(student)
```

```python
import pandas as pd
import sqlite3

connection = sqlite3.connect("meine_datenbank.db")

# SQL-Abfrage direkt in DataFrame laden!
df = pd.read_sql_query("SELECT * FROM studenten", connection)
print(df)
print("---")
```

```python
# komplexe Abfrage direkt in DataFrame
df_komplex = pd.read_sql_query("""
    SELECT vorname, nachname, note
    FROM studenten
    WHERE note IS NOT NULL
    ORDER BY note ASC
""", connection)
print(df_komplex)
print("---")
```

```python
# dann Pandas-Methoden drauf anwenden!
print(f"Durchschnittsnote: {df['note'].mean():.2f}")
print(f"Beste Note: {df['note'].min()}")
print(f"Schlechteste Note: {df['note'].max()}")
```

```python
# Grundlegende Filterung und Aggregation → in SQL
cursor.execute("SELECT AVG(note) FROM studenten WHERE aktiv = 1")

# Komplexe Analyse, Visualisierung → in Pandas
df = pd.read_sql_query("SELECT * FROM studenten", connection)
df.groupby("aktiv")["note"].mean()
```

```python
# with schließt die Verbindung automatisch — auch bei Fehlern!
with sqlite3.connect("meine_datenbank.db") as connection:
    cursor = connection.cursor()
    cursor.execute("SELECT * FROM studenten")
    for student in cursor.fetchall():
        print(student)
# hier ist die Verbindung bereits automatisch geschlossen!
```
