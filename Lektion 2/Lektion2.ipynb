```python
meine_liste = [1, 2, 3]
mein_iter = iter(meine_liste)

# Liste ist iterierbar aber KEIN Iterator
print(hasattr(meine_liste, "__next__"))  # False!

# Iterator HAT __next__
print(hasattr(mein_iter, "__next__"))    # True!
```

```python
# normale Funktion — berechnet alles auf einmal
def addiere_eins(liste):
    ergebnis = []
    for i in liste:
        ergebnis.append(i+1)
    return ergebnis

print(addiere_eins(meine_liste))  # [2, 3, 4]

# Generator — berechnet nur das nächste Element
def addiere_eins_generator(liste):
    for i in liste:
        yield i + 1

meine_liste = [0,3,6,9,12]

# normale Funktion
mein_fun = addiere_eins(meine_liste)
print(addiere_eins(meine_liste))        # [1, 4, 7, 10, 13]
print(mein_fun)

# Generator
mein_gen = addiere_eins_generator(meine_liste)

print(mein_gen)
print(next(mein_gen))
print(next(mein_gen))


```

```python
import itertools

# chain - listen hintereinander

liste1 = [1,2,3]
liste2 = [4,5,6]
for zahl in itertools.chain(liste1,liste2):
    print(zahl)
```

```python
# Product alle Kombinationen
farben = ["rot", "blau"]
groessen = ["S","M","L"]
for kombi in itertools.product(farben,groessen):
    print(kombi)
```

```python
# Combinations - alle Kombos ohne Wiederholung
buchstaben = ["A","B","C","D","E"]
for kombi in itertools.combinations(buchstaben,5):
    print(kombi)

```

```python
import itertools

# count() — zählt ab einem Startpunkt mit einem definierten Schritt

mein_iter = itertools.count(start=42, step=3)
print(next(mein_iter))
print(next(mein_iter))
print(next(mein_iter))


```

```python
# cycle() — wiederholt eine Liste endlos
mein_iter = itertools.cycle(["Michelle", "Jonas"]) 

print(next(mein_iter))   # Michelle
print(next(mein_iter))   # Jonas
print(next(mein_iter))   # Michelle — fängt wieder von vorne an
```

```python
# repeat() — wiederholt ein Element eine bestimmte Anzahl
mein_iter = itertools.repeat("Hallo!", times=3)
print(list(mein_iter))   # ['Hallo!', 'Hallo!', 'Hallo!']
```
