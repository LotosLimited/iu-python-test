```python
def berechne_bmi (gewicht, groesse):
    """
    Berechnet den BMI einer Person.
    input arguments:
    - gewicht: float (in kg)
    - groesse: float (in Meter)
    output:
    - BMI als float
    """
    bmi = gewicht / (groesse ** 2)
    return bmi
print(berechne_bmi(70, 1.75))

help(berechne_bmi)  
```

```python
def begruesse(name, sprache="deutsch"):
    """
    Begrüßt eine Person in einer bestimmten Sprache.
    input arguments:
    - name: str
    - sprache: str (Standard: Deutsch)
    output:
    - Begrüßungstext als str
    """
    if sprache == "deutsch":
        return f"Hallo {name}!"
    elif sprache == "englisch":
        return f"Hello {name}!"
    else:
        return f"Hola {name}!"
    
print(begruesse("Michelle"))
print(begruesse("Jonas", "englisch"))
print(begruesse("Maria", "ee"))
```

```python
# *args — beliebig viele Parameter als Liste
def summiere(*args):
    ergebnis = sum(args)
    print(ergebnis)

summiere(1, 2, 3)        # 6
summiere(10, 20)         # 30
summiere(5)              # 5

# **kwargs — beliebig viele Parameter als Dictionary
def summiere_dict(**kwargs):
    ergebnis = sum(kwargs.values())
    return ergebnis

print(summiere_dict(a=3, b=7, c=9))   # 19
```

```python
def äußere():
    x = "äußerer Wert"
    
    def innere():
        nonlocal x
        x = "innerer Wert"
    
    innere()
    return x

print(äußere())  # was kommt raus?
```
